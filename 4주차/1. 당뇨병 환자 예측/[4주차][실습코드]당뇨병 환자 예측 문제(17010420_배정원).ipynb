```python
import torch
import torch.optim as optim

torch.manual_seed(1)
```

```python
#데이터 가져오기
import numpy as np 
import pandas as pd

train_data=pd.read_csv('train.csv')
train_data
```

```python
#데이터 가공
x_train=train_data.copy()
x_train=x_train.drop('Diabetes',axis=1)
x_train = torch.FloatTensor(x_train.values)


y_train=train_data['Diabetes'].copy()
y_train = torch.FloatTensor(y_train.values)
y_train=y_train.reshape(537,1)

print(x_train)
print(y_train)
```

```python
# 모델 초기화
W = torch.zeros((8, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W, b], lr=1e-4)

nb_epochs = 1000000
for epoch in range(nb_epochs + 1):

    hypothesis = torch.sigmoid(x_train.matmul(W) + b)
    loss=torch.nn.BCELoss()
    cost=loss(hypothesis,y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

```python
#데이터 평가
hypothesis = torch.sigmoid(x_train.matmul(W) + b)

prediction = hypothesis >= torch.FloatTensor([0.5])

correct_prediction = prediction.float() == y_train
print(correct_prediction[:5])

accuracy = correct_prediction.sum().item() / len(correct_prediction)
print('The model has an accuracy of {:2.2f}% for the training set.'.format(accuracy * 100))
```

```python
#테스트 데이터 가져오기
test_data=pd.read_csv('test.csv')
test_data
```

```python
#정확도 측정
test_data = torch.FloatTensor(test_data.values)
hypothesis = torch.sigmoid(test_data.matmul(W) + b)
prediction = hypothesis >= torch.FloatTensor([0.5])
print(prediction)
```

```python
id=[]
diabetes=[]

for i in range(len(prediction)):
    id.append(i)
    diabetes.append(int(prediction[i].item()))
```

```python
submit = pd.DataFrame({'Id':id,
                       'Diabetes':diabetes})
submit.to_csv("submit.csv",index=None)
```
