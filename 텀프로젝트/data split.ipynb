```python
#데이터 가져오기
import numpy as np 
import pandas as pd

dataset = pd.DataFrame(pd.read_csv('도로종류별_가해자_차종별_교통사고_20221105224643.csv'))
dataset
```

```python
dataset_shuffled=dataset.sample(frac=1).reset_index(drop=True)
dataset_shuffled
```

```python
train_data=dataset_shuffled.iloc[:630,:]
test_data=dataset_shuffled.iloc[630:,:]
print(train_data.shape)
print(test_data.shape)
```

```python
# csv 변환
train_data.to_csv('train.csv',index=False)
test_x=test_data.iloc[:,:-1]
test_y=test_data.iloc[:,-1]
test_x.to_csv('test.csv',index=False)
test_y.to_csv('answer.csv',index=False)
```
