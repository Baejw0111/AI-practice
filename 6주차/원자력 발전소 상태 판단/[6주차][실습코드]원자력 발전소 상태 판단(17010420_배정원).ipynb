```python
import torch
import torch.optim as optim

#데이터 가져오기
import numpy as np 
import pandas as pd

x_train=pd.read_csv('pca_train.csv')
y_train=pd.read_csv('train_label.csv')
x_train
```

```python
y_train
```

```python
#데이터 표준화
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

x_train=scaler.fit_transform(x_train)
```

```python
x_train = torch.FloatTensor(x_train)

y_train=y_train['label'].copy()
y_train = torch.LongTensor(y_train)

print(x_train)
print(y_train)
```

```python
print(y_train.max())
```

```python
## 모델 설계
linear1 = torch.nn.Linear(256, 198, bias=True)
linear2 = torch.nn.Linear(198, 198, bias=True)
linear3 = torch.nn.Linear(198, 198, bias=True)
linear4 = torch.nn.Linear(198, 198)
# sigmoid = torch.nn.Sigmoid()
```

```python
model = torch.nn.Sequential(linear1, linear2, linear3, linear4)
```

```python
loss = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for stop in range(1500):
    
    # 그래디언트 초기화
    optimizer.zero_grad()
    # Forward 계산
    hypothesis = model(x_train)
    # Error 계산
    cost = loss(hypothesis, y_train)
    # Backward 계산 
    cost.backward()
    # 가중치 갱신
    optimizer.step()

    if stop % 100 == 0:
        print(stop, cost.item())
```

```python
## w,b 평가

with torch.no_grad(): # 임시로 required_grad = false로 설정하는 것과 같다.

    hypothesis = model(x_train)
    predicted = torch.max(hypothesis,1)
    accuracy = (predicted.indices == y_train).float().mean()
    print('\n Hypothesis: ', hypothesis, '\n Correct: ', predicted, '\n Accuracy: ', accuracy.item())
```

```python
#테스트 데이터 가져오기
test_data=pd.read_csv('pca_test.csv')
test_data
```

```python
test_data=scaler.transform(test_data)
test_data=torch.FloatTensor(test_data)
predict=model(test_data)
```

```python
submit=pd.read_csv("sample_submission.csv")

predicted = torch.max(predict,1)

for i, value in enumerate(predicted.indices):
    submit["label"][i]=value.item()

submit.to_csv("submit_1.csv",index=False)
```
